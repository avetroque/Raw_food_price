```python
# If not already installed, do: pip install pandas fastparquet
import pandas as pd

URL_DATA = 'https://storage.googleapis.com/dosm-public-economy/salaries_age_sex.parquet'

df = pd.read_parquet(URL_DATA)
if 'date' in df.columns: df['date'] = pd.to_datetime(df['date'])

df.head()
```

```python
df["sex"].unique()
```

```python
df[df["variable_en"]=='20-24'].plot.bar('year','median')
```

```python
#df.query('variable_en=="20-24" and sex=="male"').plot.bar('year','median')
#df.query('variable_en=="20-24" and sex=="female"').plot.bar('year','median')


from matplotlib import pyplot as plt

dfcopy = df.copy()

male_df = dfcopy.query('variable_en=="20-24" and sex=="male"')
female_df = dfcopy.query('variable_en=="20-24" and sex=="female"')



ax = male_df.plot(x='year', y='median',color={'blue'})
female_df.plot(ax=ax, x='year', y='median',color='red')
plt.legend(['male median','female median'])
plt.show()


male_df.head()
```
